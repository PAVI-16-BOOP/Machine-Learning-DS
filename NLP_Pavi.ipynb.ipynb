### Data Preprocessing

### removing non-useful words from the dataframe 

```python
import string
```

```python
mess = "Sample message ! Notice: It has punctuation."
```

```python
nonpunc = [c for c in mess if c not in string.punctuation]
nonpunc
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words("english")[0:20]
```

```python
nonpunc = ''.join(nonpunc)
```

```python
nonpunc
```

```python
nonpunc.split()
```

```python
clean_mess = [word for word in nonpunc.split() if word.lower() not in stopwords.words("english") ]
clean_mess
```

```python
def txt_process(msg):
    """ 1. remove punctuations
        2. remove stopwords
        3. return the cleaned string
    """
    nonpunc = [ch for ch in msg if ch not in string.punctuation]
    
    nonpunc = ''.join(nonpunc)
    
    return  [word for word in nonpunc.split() if word.lower() not in stopwords.words("english") ]
        
```

```python
messages.head()
```

```python
messages['message'].head(5).apply(txt_process)
```

### Vectorization : creates a sparse matrix where rows = different sentences in the document and columns = different words across the messages

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_transformer = CountVectorizer(analyzer = txt_process).fit(messages['message'])
```

```python
print(len(bow_transformer.vocabulary_)) # no of different words in the documents acorss all the messages
```

```python
mess4 = messages['message'].iloc[3]
mess4
```

```python
# example usage

bow4 = bow_transformer.transform([mess4])
print(bow4)
```

```python
print(bow4.shape)
```

```python
bow_transformer.get_feature_names_out()[4068]
```

```python
bow_transformer.get_feature_names_out()[9554]
```

```python
message_bow = bow_transformer.transform(messages["message"])
```

```python
print('Shape of the Sparse Matrix :' ,message_bow.shape)
```

```python
message_bow.nnz
```

```python
sparsity = (100.0 * message_bow.nnz / (message_bow.shape[0] * message_bow.shape[1]))
print('sparsity: {}'.format(round(sparsity)))
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer = TfidfTransformer().fit(message_bow)
tfidf4 = tfidf_transformer.transform(bow4)
print(tfidf4)
```

```python
# finding idf values for 2 words 'u' and 'university'

print(tfidf_transformer.idf_[bow_transformer.vocabulary_['u']])
print(tfidf_transformer.idf_[bow_transformer.vocabulary_['university']])
```

```python
messages_tfidf = tfidf_transformer.transform(message_bow)
print(messages_tfidf.shape)
```

### Training the Model through Naive Bayes Algorithm

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect = MultinomialNB().fit(messages_tfidf , messages['label'] )
```

```python
# using the model to predict the label for the 4th message in the document 

spam_detect.predict(tfidf4)[0]
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
msg_train ,msg_test ,label_train ,label_test = train_test_split(messages['message'] , messages['label'] ,test_size = 0.3 ,random_state = 101)
```

### using sklearn's pipeline to do all the preprocessing at one go for us

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('vectorizer' , CountVectorizer(analyzer = txt_process)) , 
    ("tfidf" , TfidfTransformer()) ,
    ("classifier" , MultinomialNB())
])
```

```python
pipeline.fit(msg_train , label_train)
```

```python
pred = pipeline.predict(msg_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(label_test , pred))
```
