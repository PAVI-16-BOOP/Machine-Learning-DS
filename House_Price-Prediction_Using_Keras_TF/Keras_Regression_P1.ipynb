### making the  model

```python
x=df.drop('price' , axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### **  *we get an estimate of the number of neurons from the number of features in our dataset* **

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
# smaller the batch_size = lesser the overfitting
model.fit(x = x_train,y = y_train,validation_data = (x_test,y_test), batch_size=128, epochs=360,verbose=20)
```

```python
val_loss = model.history.history['val_loss']

# Get the epoch number (0-indexed) where val_loss is minimum
best_epoch = np.argmin(val_loss)
best_epoch
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error ,explained_variance_score
```

```python
pred = model.predict(x_test)
```

```python
mean_squared_error(y_test , pred)**0.5
```

```python
 mean_absolute_error(y_test , pred)
```

```python
df['price'].describe()
```

```python
# our mean 'price'is 540296.6 but our mean absolute error is about 103451.049
# so we can improve
```
